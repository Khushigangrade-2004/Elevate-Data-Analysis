```python
# Importing Requirements

#data manipulation
import pandas as pd

#data visulization
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
# IMPORTING THE DATASATS

df = pd.read_csv("superstore_salesCSV.csv", encoding='latin1')

```

```python
#starting rows of dataset
df.head(20)
```

```python
#last rows of dataset

df.tail()
```

```python
#groupby sales by product name

sum_of_product = df.groupby("product_name")["sales"].sum()

#groupby profit by region
revenue_by_region = df.groupby("region")["profit"].sum()

```

```python
#sumsof  sales by state
sale_by_state = df.groupby("state")["sales"].sum()
print(sum_of_product)
print(revenue_by_region)
print(sale_by_state)
```

```python
#shows plot

revenue_by_region.plot(kind='bar', title='Total Profit by Region', figsize=(12, 4), color='green')
plt.ylabel("Profit")
plt.xlabel("Region")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
